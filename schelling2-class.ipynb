```python
import matplotlib.pyplot as plt
import itertools as it
import random
```

```python
class Schelling:
    """Schelling's segregation model"""
    version = 0.2  # class attribute
    
    def __init__(self, width, height, empty_ratio = 0.25):
        """constructor"""
        self.width = width
        self.height = height
        self.empty_ratio = empty_ratio  # class instance attribute
        print("Created a class instance with empty ratio", empty_ratio)
        
    def populate(self):
        """The populate method is used at the beginning of the simulation. This method randomly distributes people in the grid."""
        #self.all_houses = [(0, 0), (1, 0), (0, 1), (1, 1)]
        self.all_houses = list(it.product(range(self.width), range(self.height)))
        random.shuffle(self.all_houses)
        
        self.n_empty = int(self.empty_ratio * len(self.all_houses))
        self.empty_houses = self.all_houses[:self.n_empty]
        
        self.remaining_houses = self.all_houses[self.n_empty:]
        
        houses_by_color = []
        for house in self.remaining_houses:
            #houses_by_color.append((house, 1))
            if (house[0] + house[1]) % 2 > 0:
                color = 1
            else:
                color = 2
            houses_by_color.append((house, color))
        
        self.agents = dict(houses_by_color)
        
    def plot(self, title, file_name):
        fig, ax = plt.subplots(figsize=(5, 5))
        ax.grid(True)
        agent_colors = {1: 'b', 2: 'r'}
        for agent in self.agents:
            ax.scatter(agent[0] + 0.5, agent[1] + 0.5, color=agent_colors[self.agents[agent]])
            
        ax.set_title(title, fontsize=10, fontweight='bold')
        ax.set_xlim([0, self.width])
        ax.set_ylim([0, self.height])
        plt.savefig(file_name)
```

```python
schelling_1 = Schelling(8, 8)
schelling_1.populate()
print(schelling_1.agents)
```

```python
schelling_1.plot('Schelling Model: Initial State', 'schelling_initial.png')
```

```python
models = []
for empty_ratio in [0.25, 0.5, 0.75]:
    schelling = Schelling(8, 8, empty_ratio)
    models.append(schelling)
```

```python
for model in models:
    model.populate()
    print(model.agents)
```

```python
for model in models:
    model.plot('Schelling Model: Initial State, E-R = ' + str(model.empty_ratio), 'schelling_initial.png')
```
